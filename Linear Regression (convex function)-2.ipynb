```python
import tensorflow as t
import matplotlib.pyplot as plt
```

```python
x_data = [1,2,3]
y_data = [1,2,3]

#W = t.Variable(t.random_normal([1]), name='weight')
W = t.Variable(5.0)
X = t.placeholder(t.float32)
Y = t.placeholder(t.float32)

hypothesis = X * W

cost = t.reduce_mean(t.square(hypothesis - Y))
```

```python
'''
learning_rate = 0.1
gradient = t.reduce_mean((W * X - Y) * X)
descent = W - learning_rate * gradient
update = W.assign(descent)

'''

optimazer = t.train.GradientDescentOptimizer(0.1)
train = optimazer.minimize(cost)

```

```python
sess = t.Session()

sess.run(t.global_variables_initializer())
```
