```python
import tensorflow as t
import tensorflow as tf
```

```python
x_data = [
    [1., 2., 3., 4., 5.],
    [73., 93., 89., 96., 73.],
    [80., 88., 91., 98., 66.],
    [75., 93., 90., 100., 70.]
]

y_data = [[50.],
          [185.],
          [180.], 
          [196.]]
```

```python
X = t.placeholder(t.float32, shape=[None, 5])
Y = t.placeholder(t.float32, shape=[None, 1])
```

```python
W = t.Variable(t.random_normal([5, 1]), name = 'weight')
#W = t.placeholder(t.float32, shape=[5, 1])
b = t.Variable(t.random_normal([1]), name = 'bias')

hypothesis = t.matmul(X, W) + b
cost = t.reduce_mean(t.square(hypothesis - Y))
```

```python
sess = t.Session()
sess.run(t.global_variables_initializer())
print(sess.run([hypothesis], feed_dict={
    X: x_data, W: [[1],[1],[1],[1],[1]]
}))

test = sess.run(cost, feed_dict={
        X: x_data, W: [[1],[1],[1],[1],[1]],Y:y_data})

print(test)
```

```python
optimizer = t.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost)
```

```python
sess = t.Session()

sess.run(t.global_variables_initializer())

for i in range(2001):
    print(sess.run([cost, hypothesis, train],
            feed_dict={
                 X: x_data,
                 Y: y_data
    }))
```
