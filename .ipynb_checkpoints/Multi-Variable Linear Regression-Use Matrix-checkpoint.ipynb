```python
import tensorflow as t
import tensorflow as tf
```

```python
x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]

y_data = [152., 185., 180., 196., 142.]
```

```python
x1 = t.placeholder(t.float32)
x2 = t.placeholder(t.float32)
x3 = t.placeholder(t.float32)
Y = t.placeholder(t.float32)
```

```python
w1 = f.Variable(t.random_normal([1]), name = 'weight1')
w2 = f.Variable(t.random_normal([1]), name = 'weight2')
w3 = f.Variable(t.random_normal([1]), name = 'weight3')
b = f.Variable(t.random_normal([1]), name = 'bias')

hypothesis = (x1 * w1) + (x2 * w2) + (x3 * w3)
```

```python
cost = t.reduce_mean(t.square(hypothesis - Y))

optimizer = t.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost)
```

```python
sess = t.Session()

sess.run(t.global_variables_initializer())


sess.run(train,
         feed_dict={
             x1: x1_data,
             x2: x2_data,
             x3: x3_data, 
             Y: y_data})

print(sess.run([cost, hypothesis, train],
        feed_dict={x1: x1_data, x2: x2_data, x3: x3_data, Y: y_data})
)
```

```python
for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
        feed_dict={x1: x1_data, x2: x2_data, x3: x3_data, Y: y_data})
    if step % 10 == 0:
        print(step, "Cost: ", cost_val, "\nPrediction:\n", hy_val)
        
```
